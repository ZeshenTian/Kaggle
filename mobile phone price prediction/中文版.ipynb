首先检查每一列是否有数据缺失

```python
# 检查每一列是否存在缺失数据并打印出存在缺失数据的列名
for column in data.columns:
    if data[column].isnull().any():
        print(f"列 '{column}' 中存在缺失数据")
```

对于所有的object类型数据，必须经过处理才能输入回归模型中。我们先探测每个值为object类型的列有多少类别，以确认如何对每一列进行处理

```python
# 获取从第三列开始到最后一列结束的所有列
subset_data = data.iloc[:, 2:]

# 计算每列的唯一值数量
unique_counts = subset_data.apply(pd.Series.nunique)

# 输出列名和类别数量
for column, count in unique_counts.items():
    print(f"{column}: {count}")
```

1. 对No_of_sim这列特征进行处理。由于特征数量较小，所以首先查看一下这列特征的所有值

```python
unique_styles = data['No_of_sim'].unique()
for style in unique_styles:
    print(style)
```

这其实是六个特征 是否dual sim 是否3G 是否4G 是否5G，是否VoLTE,是否Vo5G。考虑将它拆分为六个布尔型特征。

```python
# 选择要拆分的列
column_to_split = 'No_of_sim'

# 创建新列
data['Dual Sim'] = data[column_to_split].str.contains('Dual Sim').astype(int)
data['3G'] = data[column_to_split].str.contains('3G').astype(int)
data['4G'] = data[column_to_split].str.contains('4G').astype(int)
data['5G'] = data[column_to_split].str.contains('5G').astype(int)
data['Volte'] = data[column_to_split].str.contains('VoLTE').astype(int)
data['Vo5G'] = data[column_to_split].str.contains('Vo5G').astype(int)
# 删除原始列
data.drop(columns=[column_to_split], inplace=True)
# 打印处理后的数据
print(data.head())
```

```python
# 选择你要计算的几列
columns_to_sum = ['Dual Sim', '3G', '4G', '5G', 'Volte', 'Vo5G']

# 分别对每一列进行求和并输出结果
for column in columns_to_sum:
    column_percent = data[column].sum() / data.shape[0]
    print(f"有 '{column}' 的手机的比例为: {column_percent:.4f}")
```

2. 对RAM的特征进行处理

```python
unique_styles = data['Ram'].unique()
for style in unique_styles:
    print(style)
```

我们发现数据存在乱码现象，比如：128 GB inbuilt 256 GB inbuilt 512 GB inbuilt是Inbuilt_memory列的数据，Helio G90T是Processor_name的数据，因此需要先对这几行数据进行处理。先显示存在异常数据的相应行的有关字段

```python
exception_values = ['128 GB inbuilt','6000 mAh Battery with 22.5W Fast Charging','256 GB inbuilt',
                    '512 GB inbuilt','Helio G90T']
```

```python
data[data['Ram'].isin(exception_values)].loc[:, ['Ram', 'Inbuilt_memory', 'Processor_name','Processor','Battery']]
```

我们发现对于数据1246，它的许多属性颠倒了

```python
data.loc[1246, [
    'Ram', 'Inbuilt_memory', 'External_Memory',
    'Camera', 'Display', 'Battery', 'fast_charging', 'Screen_resolution', 'Processor', 'Processor_name'
]] = [
    data.loc[1246, 'Processor_name'], data.loc[1246, 'Processor'],
    data.loc[1246, 'Camera'], data.loc[1246, 'Display'], '6.75 inches',
    '6000 mAh Battery', '22.5W Fast Charging', '720 x 1600 px Display with Water Drop Notch', np.nan, np.nan
]
```

```python
data.iloc[1246]
```

对于数据264，许多内容都颠倒了，我们进行相应替换

```python
data.loc[264, [
    'Ram', 'Battery', 'Display', 'Camera', 'External_Memory', 'Inbuilt_memory', 'fast_charging', 'Screen_resolution', 'Processor', 'Processor_name' 
]] = [
    '6 GB RAM', '4300 mAh Battery','6.5 inches', 
    data.loc[264, 'External_Memory'], 'Android', '64 GB inbuilt', '30W Fast Charging', '1080 x 2400 px', data.loc[264, 'Inbuilt_memory'], np.nan
]
```

```python
data.iloc[264]
```

对于其它数据，它的Inbuilt_memory属性空缺。应该把它们的Ram属性与Inbuilt_memory属性调换

```python
# 定义需要修改的条件
conditions = data['Ram'].isin(['128 GB inbuilt', '256 GB inbuilt', '512 GB inbuilt'])

# 将符合条件的行的 Inbuilt_memory 列赋值为 Ram 列的值
data.loc[conditions, 'Inbuilt_memory'] = data.loc[conditions, 'Ram']

# 将符合条件的行的 Ram 列赋值为 NaN
data.loc[conditions, 'Ram'] = np.nan
```

```python
unique_styles = data['Ram'].unique()
for style in unique_styles:
    print(style)
```

```python
# 提取 RAM 值列中的数值部分
data['Ram'] = data['Ram'].str.extract(r'(\d+\.?\d*)').astype(float)

# 计算 RAM 值列的中位数
ram_median = data['Ram'].median()

# 用中位数填充 RAM 值列中的 NA 值
data['Ram'].fillna(ram_median, inplace=True)
```

至此，完成对Ram特征的处理

3. 对于battery特征的处理

```python
unique_styles = data['Battery'].unique()
for style in unique_styles:
    print(style)
```

```python
# 提取 Battery列中的数值部分
data['Battery'] = data['Battery'].str.extract(r'(\d+\.?\d*)').astype(float)
```

4. 提取display列中的数值部分

```python
unique_styles = data['Display'].unique()
for style in unique_styles:
    print(style)
```

```python
data['Display'] = data['Display'].str.extract(r'(\d+\.?\d*)').astype(float)
```

5. 对camera列进行特征处理，这里采取的处理方式是将所有数字提取出来之后相加（因为实在是不好处理所以只能这样...)

```python
unique_styles = data['Camera'].unique()
for style in unique_styles:
    print(style)
```

```python
# 提取所有数字并相加
sum_values = []
for index, row in data.iterrows():
    numbers = pd.to_numeric(pd.Series(row['Camera']).str.extractall(r'(\d+)')[0], errors='coerce')
    sum_value = numbers.sum()
    sum_values.append(sum_value)

# 替换原始数据
data['Camera'] = sum_values
```

6.处理external_memory，这里选择直接删除这个特征（因为是传统机器学习不打算用深度学习的方法所以这个特征其实是很难处理的）

```python
unique_styles = data['External_Memory'].unique()
for style in unique_styles:
    print(style)
```

```python
data = data.drop(columns=['External_Memory'])
```

处理Android Version特征，选择将na值用中位数填充，7.1.1值用7.11替换，其它值直接用数值型特征

```python
unique_styles = data['Android_version'].unique()
for style in unique_styles:
    print(style)
```

```python
# 提取 Android_version 值列中的数值部分
data['Android_version'] = data['Android_version'].str.extract(r'(\d+\.?\d*)').astype(float)

# 计算 RAM 值列的中位数
ram_median = data['Android_version'].median()

# 用中位数填充 RAM 值列中的 NA 值
data['Android_version'].fillna(ram_median, inplace=True)
```

处理价格特征：直接转成整数

```python
data['Price'] = data['Price'].str.replace(',', '').astype(float)
```

```python
unique_styles = data['company'].unique()
for style in unique_styles:
    print(style)
```

处理公司特征：用one-hot编码

```python
data['company'] = data['company'].str.lower()
```

```python
# 使用 get_dummies() 函数进行编码
one_hot_encoded = pd.get_dummies(data['company'])

# 可选：将编码后的结果合并到原始数据中
data = pd.concat([data, one_hot_encoded], axis=1)
```

```python
data.drop(columns=['company'], inplace=True)
```

处理Inbuilt_memory特征：先将TB转为GB再转为数字

```python
unique_styles = data['Inbuilt_memory'].unique()
for style in unique_styles:
    print(style)
```

```python
data['Inbuilt_memory'] = data['Inbuilt_memory'].replace('1TB', '1024GB')
```

```python
data['Inbuilt_memory'] = data['Inbuilt_memory'].str.extract(r'(\d+\.?\d*)').astype(float)
```

处理fast_charging特征：直接转为数字，NA值、只写了Fast Charging的值均用0填充

```python
data['fast_charging'] = data['fast_charging'].str.extract(r'(\d+\.?\d*)').astype(float).fillna(0)
```

处理Screen_resolution特征：将数字相加，如果没有数字就用0填充

```python
data['Screen_resolution'] = data['Screen_resolution'].str.extract(r'(\d+\.?\d*)').astype(float).fillna(0)
```

处理Processor特征：

```python
unique_styles = data['Processor'].unique()
for style in unique_styles:
    print(style)
```

```python
data['Processor'] = data['Processor'].str.strip()
data['Processor'] = data['Processor'].str.replace(' Processor', '', regex=False)
data['Processor'] = data['Processor'].replace(['Nine-Cores', 'Nine Core', 'Nine Cores'], 'Nine Core')
data['Processor'] = data['Processor'].fillna('Unknown')
```

```python
unique_styles = data['Processor'].unique()
for style in unique_styles:
    print(style)
```

```python
one_hot_encoded = pd.get_dummies(data['Processor'])
data = pd.concat([data, one_hot_encoded], axis=1)
```

```python
data.drop(columns=['Processor'], inplace=True)
```

处理Processor_name：删除该特征

```python
data.drop(columns=['Processor_name'], inplace=True)
```
